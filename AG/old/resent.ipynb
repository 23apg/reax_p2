```python
ssho = []
#state level
for state in states:
    s = state.lower()
    docs = db.reddit_state.aggregate(
    [ { "$match" : { 'location' : s } } ]
    )
    vsc = []
    vsp = []
    vsn = []
    kc = []
    ku = [] 
    print(state)
#     print(state)
#     print(len(list(docs)))
    doclength = 0

    if (len(list(docs)) != doclength):
        print("true -----------------------")
        docs = db.reddit_state.aggregate([ { "$match" : { 'location' : s } } ])
#     elif (s == 'nevada'):
#         print("nevada")
#         docs = db.reddit_state.aggregate([ { "$match" : { 'location' : state } } ])
    else:
        print("false______________________________")
        docs = db.reddit_state.aggregate([ { "$match" : { 'location' : state } } ])

#document of state level
    for d in docs:
        #if docs level
        if (len(d['comment_dict']) > 0): 
            length = len(d['comment_dict'])
            
            #for each d in docs
            for x in range(length):
                vsc.append(d['comment_dict'][x]['vader']['compound'])
                vsp.append(d['comment_dict'][x]['vader']['pos'])
                vsn.append(d['comment_dict'][x]['vader']['neg'])
                kc.append(d['comment_dict'][x]['karma']['compound'])
                ku.append(d['comment_dict'][x]['karma']['ups'])
#             print(len(vsc))
#             print(len(vsp))
#             print(len(vsn))
#             print(len(kc))
#             print(len(ku))
        
    
    cs = float(np.mean(vsc))
    pos = float(np.mean(vsp))
    neg = float(np.mean(vsn))
    kcom = float(np.sum(kc))
    kup = float(np.sum(ku))
#     print(cs)
#     print(pos)
#     print(neg)
#     print(kcom)
#     print(kup)
    day = dt.datetime.now().strftime('%Y-%m-%d')
    
    location_sentiments = dict( location = s,
                                datetime = day,
                                vaderCom = cs,
                                vaderPos = pos,
                                vaderNeg = neg,
                                karmaUp = kup,
                                karmaCom = kcom )
#     print(location_sentiments)
    
    ssho.append(location_sentiments)
    
    
    
       
        
            
                
```

```python
csho = []
#state level
for city in city_subreddits:
    c = city.lower()
    docs = db.reddit_city.aggregate(
    [ { "$match" : { 'location' : city, 'datetime': dt.datetime.now().strftime('%Y-%m-%d') } } ]
    )
    vsc = []
    vsp = []
    vsn = []
    kc = []
    ku = [] 
    print(city)
    

    doclength = 0

    if (len(list(docs)) != doclength):
        print("true -----------------------")
        docs = db.reddit_city.aggregate([ { "$match" : { 'location' : city, 'datetime': dt.datetime.now().strftime('%Y-%m-%d') } } ])

    else:
        print("false______________________________")
        docs = db.reddit_city.aggregate([ { "$match" : { 'location' : c, 'datetime': dt.datetime.now().strftime('%Y-%m-%d') } } ])
        
#document of state level
    for d in docs:
        #if docs level
        if (len(d['comment_dict']) > 0): 
            length = len(d['comment_dict'])
            
            #for each d in docs
            for x in range(length):
                vsc.append(d['comment_dict'][x]['vader']['compound'])
                vsp.append(d['comment_dict'][x]['vader']['pos'])
                vsn.append(d['comment_dict'][x]['vader']['neg'])
                kc.append(d['comment_dict'][x]['karma']['compound'])
                ku.append(d['comment_dict'][x]['karma']['ups'])
#             print(len(vsc))
#             print(len(vsp))
#             print(len(vsn))
#             print(len(kc))
#             print(len(ku))
        
    
    cs = float(np.mean(vsc))
    pos = float(np.mean(vsp))
    neg = float(np.mean(vsn))
    kcom = float(np.sum(kc))
    kup = float(np.sum(ku))
#     print(cs)
#     print(pos)
#     print(neg)
#     print(kcom)
    print(kup)
    
    if (c == 'nyc'):
        c = "newyorkcity"
    elif (c == 'twincitiessocial'):
        c = 'minneapolis'
    elif (c == 'bayarea'):
        c = 'oakland'
    else:
        c = c
        
    day = dt.datetime.now().strftime('%Y-%m-%d')
    
    location_sentiments = dict( location = c,
                                datetime = day,
                                vaderCom = cs,
                                vaderPos = pos,
                                vaderNeg = neg,
                                karmaUp = kup,
                                karmaCom = kcom )
#     print(location_sentiments)
    
    csho.append(location_sentiments)

```
