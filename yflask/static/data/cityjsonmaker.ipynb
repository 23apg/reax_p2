```python
import json
from pprint import pprint
import numpy as np
import datetime as dt
import pymongo
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.project_reax
```

```python
docs = db.city_full_lite.find({})
```

```python
holder = []

for d in docs:
    citycity = d['city']
    city = citycity.replace(' ', '')
    latlong = [d['latitude'], d['longitude']]
    
    cityll = {'city' : citycity,
              'll' : latlong,
             }
    holder.append(cityll)
```

```python
reddocs = db.city_simple2.find({})
```

```python
empty = []
cities = []


for r in reddocs:
    rcity = r['location']
    rclowernospace = rcity.lower().replace(' ','')
    ph = r['percposhead']
    nh = r['percneghead']
    neuh = r['percneuhead']
    pc = r['posavgcomms']
    nc = r['negavgcomms']
    neuc = r['neuavgcomms']
    kp = r['avgupscorePOSHL']
    kn = r['avgupscoreNEGHL']
    kn = r['avgupscoreNEUHL']
    wc = r['wordcloud']
    
    for h in holder:
        city = h['city']
        ll = h['ll']
        citylowernospace = city.lower().replace(' ','')
        if ((citylowernospace == rclowernospace) and (rclowernospace not in empty)):
            x = {}
            x['city'] = city
            x['ll'] = ll
            x['posHEADLINEperc'] = ph
            x['negHEADLINEperc'] = nh
            x['neuHEADLINEperc'] = neuh
            x['posavgcomms'] = pc
            x['negavgcomms'] = nc
            x['neuavgcomms'] = neuc
            x['avgKARMApos'] = kp
            x['avgKARMAneg'] = kn
            x['avgKARMAneu'] = kn
            x['wc'] = wc
            
            cities.append(x)
            empty.append(rclowernospace)
            
        elif ((citylowernospace == rclowernospace) and (rclowernospace in empty)):
            pass
        else:
            pass
```
